```python
import numpy as np
cov = np.cov(X, rowvar = False, bias=True)
cor = np.corrcoef(X)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(cov, annot = True)

plt.show()
```
